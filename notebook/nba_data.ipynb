This notebook aims to show how to use the functions of the 
`nba_stat_data` in a notebook context.

```python
import pandas as pd
import nba_stat_data as nba
```

## Look at the data

The module contains a data frame with data from NBA with statistics over 
the best players.

```python
nba.df_nba.head()
```

One can create bins according to the height of players.

```python
df = nba.df_nba.copy()
df = df.assign(height_bins=pd.qcut(df.height, q=4))
df.head()
```

```python
df.groupby("height_bins").count()["Year"].plot.bar()
```

## Scatter plot

The `nba_scatter` functions produce a scatter plot using two columns of 
the data frame.

```python
nba.nba_scatter(nba.df_nba, "height", "bmi")
```

## Pivot table

The `nba_pivot_table` function produce a table using the positions as
columns, the height bins as rows and the given column as averaged values.

```python
nba.nba_pivot_table(nba.df_nba, "bmi")
```
